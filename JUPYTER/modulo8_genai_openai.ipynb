# Modulo 8 – Generative AI con OpenAI

In questo notebook:

- Usiamo l'API OpenAI per generare testo
- Esempio di TTS (Text-to-Speech)
- Mini-esempio di RAG con OpenAI (solo schema)

Assicurati di avere la variabile d'ambiente `OPENAI_API_KEY` configurata.

```
from openai import OpenAI
import os

client = OpenAI()

prompt = 'Spiega in modo semplice che cos\'è la regressione lineare.'
res = client.chat.completions.create(
    model='gpt-4o-mini',
    messages=[
        {'role': 'system', 'content': 'Sei un assistente didattico.'},
        {'role': 'user', 'content': prompt},
    ],
    temperature=0.4,
)
print(res.choices[0].message.content)
```

## TTS – Text-to-Speech con OpenAI

Questo esempio salva un file MP3 con la voce generata.

```
text = 'Benvenuto al corso di Machine Learning con Python!' 
response = client.audio.speech.create(
    model='gpt-4o-mini-tts',
    voice='alloy',
    input=text,
    format='mp3',
)
audio_bytes = response.read()
with open('benvenuto_ml.mp3', 'wb') as f:
    f.write(audio_bytes)
print('File benvenuto_ml.mp3 salvato.')
```
